```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import jieba
```

```python
df=pd.read_csv('data/crawler/url_title.csv',index_col=0)
df
```

```python
place_names=pd.read_csv('data/geo/china_city_list.csv',encoding='gbk')
place_names  #有东西部等区域划分和几线城市的划分
```

```python
def extract_province_names(title):
    possible_places = jieba.lcut(title)
    province=np.append(place_names['Province'].unique(), '内蒙')
    return [place for place in possible_places if place in province]  
def extract_city_names(title):
    possible_places = jieba.lcut(title)
    city=place_names['City'].unique()
    return [place for place in possible_places if place in city]   
df['province'] = df['title'].apply(extract_province_names)
df['city'] = df['title'].apply(extract_city_names)
```

```python
df
```

```python
list_province=[]
list_city=[]
for i in range(len(df)):
    list_province+=df['province'][i]
    list_city+=df['city'][i]
province_count=pd.DataFrame(list_province).value_counts()
province_count.columns=['省份']

city_count=pd.DataFrame(list_city).value_counts()
city_count.columns=['城市']
city_count
```

```python
province_count.to_csv('data/result/province_count.csv')
city_count.to_csv('data/result/city_count.csv')
```
